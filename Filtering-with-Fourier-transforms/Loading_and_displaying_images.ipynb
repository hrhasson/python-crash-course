Now that we have our file uploaded to Colab, let's load it into python. We use cv2 to read our color image into python as a greyscale image. This takes the pixel values for each color and combines them to one single value per pixel. We will name the image "img" for now.

```
# Load an color image in grayscale
img = cv2.imread('practice_image.jpg',0)
type(img)
```

Notice that the image is just a single array of pixel values, with the position in the matrix corresponding to the pixel position in the image.

&nbsp;

To display the image, we use matplotlib (plt) like we usually do, but instead of the familiar *plt.plot()* we now use **plt.imshow()** since we have a matrix of data.

```
plt.imshow(img)
```

Notice that it gives the image a weird green color scale, even though the image is in grayscale. You can modify the color scale of pixel values with the keyword argument **cmap**. There are multiple options you can choose ([see here](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html)), but we will stick with grayscale for now:

```
plt.imshow(img, cmap = 'gray')
```
